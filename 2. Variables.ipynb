### - **Declaring And Assigning Variables**

```python
age = 29
height = 6.0
name = "Tajamul"
is_student = True

## printing the variables
print(f"age :, {age}")
print("Height:",height)
print("Name:",name)
```

### - **Naming Conventions**

- Variable names should be descriptive
- They must start with a letter or an '_' and contains letter,numbers and underscores
-  variables names case sensitive

```python
#valid variable names

first_name="KRish"
last_name="Naik"

# Invalid variable names
2age = 30
first-name = "Tajamul"
@name = "Tajamul"
```

```python
## Understnading Variable data types
age=25 #int
height=6.0 #float
name="Tajamul" #str
is_student=True #bool
```

### - **Type Checking and Type Conversion**

```python
# Type Checking and Conversion
type(height)
```

```python
# Type conversion
age='25'
print(type(int(age)))
```

### - **Dynamic Typing**

```python
## Dynamic Typing

var=10 #int
print(var,type(var))

var="Hello"
print(var,type(var))

var=3.14
print(var,type(var))
```

```python
### Simple calculator
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

sum = num1 + num2
difference = num1 - num2
product = num1 * num2
quotient = num1 / num2

print("Sum:", sum)
print("Difference:", difference)
print("Product:", product)
print("Quotient:", quotient)
```
